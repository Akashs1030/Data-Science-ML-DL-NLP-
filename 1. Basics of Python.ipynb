### 1. Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(n):
    if n <= 1:
        return False 
    if n == 2:
        return True   
    if n % 2 == 0:
        return False  
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
```

```python
num = int(input('enter a number '))
if is_prime(num):
    print(f'{num} is a prime number')
else:
    print(f'{num} is not a prime number')
```

### 2. Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random
a=random.randint(1, 10)
b=random.randint(1, 10)
print(f' what is {a} × {b}?')
user_ans=int(input('enter your answer:'))
correct_ans=a*b
if user_ans==correct_ans:
    print('correct')
else:
    print('incorrect')
```

### 4. A hospital charges its patients based on the following: • Room charges per day: ₹2,000 • Doctor consultation fee (flat): ₹1,500 • Lab test charges: Based on the number of tests taken. Each test costs ₹300 • Medicine charges: Total cost of medicines provided • Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied Write a Python program to calculate the final bill for a patient who: • Stayed for 4 days • Had 3 lab tests • Had medicine charges of ₹2,400

```python
#given
room_charge_per_day=2000
consult_fee=1500
lab_test_charge=300
disc_thresh=10000
disc_percentage=10
days_stayed=4
num_of_tests=3
med_charges=2400
```

```python
room_total=room_charge_per_day*days_stayed
lab_total=lab_test_charge*num_of_tests
```

```python
total_before_discount=room_total+consult_fee+lab_total+med_charges
```

```python
if total_before_discount > disc_thresh:
    discount=total_before_discount * disc_percentage /100
else:
    discount=0
final_bill=total_before_discount-discount
```

```python
# Final bill
print("room charges:", room_total)
print("consultation fee:", consult_fee)
print("lab test charges:", lab_total)
print("medicine charges:", med_charges)
print("total before discount:", total_before_discount)
print("discount applied:", discount)
print("final bill amount:", final_bill)
```
