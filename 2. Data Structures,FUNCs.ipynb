### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
import pandas as pd
import numpy as np
```

```python
data={'name':['akash','sanju','sam','sanju','manu','kedar','aman','kiran','abhi','varun'],
     'age':[18,19,20,19,22,23,24,25,20,21],'city':['blr','goa','pune','goa','kerala','blr','goa','pune','delhi','kerala'],
      'salary':[10000,20000,60000,20000,np.nan,50000,40000,70000,np.nan,80000]}
```

```python
df=pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df[['name','city']])
```

```python
df.drop(columns=['city'])
```

```python
df['salary'].fillna(df['salary'].mean(),inplace=True)
```

```python
df
```

```python
duplicates=df.drop_duplicates()
duplicates
```
