```python
import numpy as np
import scipy.stats as stats
```

```python
heads = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

#### a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
x_bar = np.mean(heads)
x_bar
```

```python
std_dev = np.std(heads)
std_dev
```

```python
n = 15
confidence = 0.99
alpha = 1 - confidence
### CI = 0.99, alpha = 0.01
```

```python
t_critical = stats.t.ppf(1-(alpha/2),14)
t_critical 
```

```python
marigin_of_error = t_critical * (std_dev/np.sqrt(n))
marigin_of_error
```

```python
durability_mean_lower = x_bar - marigin_of_error
durability_mean_upper = x_bar + marigin_of_error
print(durability_mean_lower ,',',durability_mean_upper)
```

#### b. Build 99% Confidence Interval Using Known Population Standard Deviation

```python
sigma = 0.2
```

```python
z_critical = stats.norm.ppf(1 - alpha/2)
z_critical 
```

```python
z_marigin_of_error = z_critical * (sigma/np.sqrt(n))
z_marigin_of_error
```

```python
z_mean_lower = x_bar - z_marigin_of_error
z_mean_upper = x_bar + z_marigin_of_error
print(z_mean_lower ,',',z_mean_upper)
```

```python
import numpy as np
import scipy.stats as stats
```

```python
mean_theor =1000 + 5 * 600
sigma = 5 * 25
n = 25
x_bar = 3050
alpha = 0.05
confidence = 1 - alpha
```

```python
z_score = (x_bar - mean_theor)/(sigma/np.sqrt(n))
z_score
```

```python
z_critical = stats.norm.ppf(confidence)
z_critical
```

```python
if z_score>z_critical:
    print(" Null Hypothesis(H0) : There is no increase in Weekly Operating Costs");
else:
    print(" Alternative Hypothesis(H1) : There is an increase in Weekly Operating Costs")
```

```python
### Calcualting P_value
p_value  = 1 - stats.norm.cdf(z_score)
p_value
```
