```python
dept_bill_summary=patients.groupby('Department')['BillAmount'].sum()
dept_bill_summary
```

```python
patients["BillAmount"]=patients["BillAmount"].fillna(patients["BillAmount"].mean())
```

```python
merged = pd.merge(patients, billing, on="PatientID")
```

```python
new_patients = pd.DataFrame({
    "PatientID": [2001, 2002],
    "Department": ["Cardiology", "Neurology"],
    "Doctor": ["Dr. Smith", "Dr. John"],
    "BillAmount": [5000, 7000],
    "Name": ["New Patient A", "New Patient B"],
    "InsuranceCovered": ["Yes", "No"],
    "FinalAmount": [4000, 7000]})
```

```python
merged = pd.concat([merged, new_patients], ignore_index=True)
```

```python
extra_col = pd.DataFrame({
    'DiscountApplied': ["Yes", "No"]*(len(merged)// 2 + 1),
    'PaymentStatus': ["Paid", "Pending"]*(len(merged)//2 + 1)
})
final_df = pd.concat([merged, extra_cols[:len(merged)]], axis=1)
```
